## CREATING DATAFRAMES

```python
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}
df=pd.DataFrame(data)
```

```python
df
```

```python
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
df_2=pd.DataFrame(data_list)
df_2
```

```python
#adding column to the dataframe of list
col = ['name','age','city','salary']
df_2.columns=col
df_2
```



1.  If you observe in dictionary pd.dataframes gives directly column wise arranged data bcoz dictionary has key ,values
2.   But when it comes to list  there are no  keys but still
we got data arranged into columns ..here the imp point to be noted is  it is not a list its lists of lists and if lists of lists come into pandas it always arranges data as it is in the list



```python
# adding columns to RAWdata which doesnt have any columns
df_3=pd.DataFrame(data_list,columns=['name','age','city','salary'])
df_3
```
